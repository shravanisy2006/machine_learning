## Why Feature Scaling?

Features in a dataset can have very different ranges.

For example:
- House Size: 852 - 2104
- Bedrooms: 2 - 5
- Floors: 1 - 2

Without feature scaling, Gradient Descent may converge slowly because features with larger values dominate the learning process.

Feature scaling transforms all features into a similar range, leading to faster and more stable convergence.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([
    [2104, 5, 1, 45],
    [1416, 3, 2, 40],
    [852, 2, 1, 35]
])
initial_w = np.zeros(X_train.shape[1])
initial_b = 0.
```

```python
m , n = X_train.shape
scaled_data =[]
for j in range (n):
    feature = X_train[: , j]
    mean = np.mean(feature).round(2)
    max = np.max(feature)
    min = np.min(feature)
    numbers=[]
    for i in feature:
        number = (i - mean ) / (max - min)
        numbers.append(number)
    scaled_data.append(numbers)
    arr = np.array(scaled_data).T
print(arr)
        
    
        
```

```python
plt.figure(figsize=(8,5))

for j in range(X_train.shape[1]):
    plt.plot(X_train[:, j], marker='o', label=f'Feature {j+1}')

plt.title("Before Feature Scaling")
plt.xlabel("Training Example")
plt.ylabel("Feature Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

for j in range(arr.shape[1]):
    plt.plot(arr[:, j], marker='o', label=f'Feature {j+1}')

plt.title("After Feature Scaling")
plt.xlabel("Training Example")
plt.ylabel("Feature Value")
plt.legend()
plt.grid(True)
plt.show()
```

## Observations

- Before scaling, the house size feature dominated the dataset.
- After scaling, all features were brought to a similar range.
- Feature scaling helps Gradient Descent converge faster and more efficiently.
