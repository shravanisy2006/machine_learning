```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0 , 2.0])
y_train = np.array([300.0 , 500.0])
```

```python
def cost (x , y , w , b):
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w *x[i] + b
        cost = (f_wb - y[i])**2
        cost_sum += cost
        total_cost = (1 / (2*m)) * cost_sum

    return total_cost
```

## Scatter Plot for prices vs size

```python
plt.scatter(x_train , y_train , c='b' , marker = 'v')
plt.title("House Prices")
plt.xlabel("size in 1000 sqft")
plt.ylabel("Prices in 1000s of dollars")
plt.grid()
plt.show()
```

## Cost vs Weight

```python
w_values = np.arange(0 , 400 , 20)
cost_values = []
for w in w_values:
    c = cost(x_train , y_train , w , 100)
    cost_values.append(c)

plt.plot(w_values , cost_values , marker = 'o')
plt.title("Cost function vs weight")
plt.xlabel("weight")
plt.ylabel("cost j(w,b)")
plt.grid()
plt.savefig("cost_vs_weight.png")
plt.show()
```

## Observations

- The cost is very high when the weight (w) is far from the optimal value.
- As w approaches 200, the cost decreases.
- The minimum cost is obtained near w = 200.
- Beyond this value, the cost starts increasing again.
- Therefore, the objective of Linear Regression is to find the parameters (w and b) that minimize the cost function.

## Different Prediction lines

```python
w_values = [50 , 100 , 200]
for w in w_values:
    predictions = w * x_train + 100 
    plt.plot(x_train , predictions , label = f'w={w}')
plt.scatter(x_train , y_train , marker = 'x' , c = 'r' , label = 'actual data')
plt.xlabel("size")
plt.ylabel("price")
plt.title('Effect of Different Values of w')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Observations

- Different values of w produce different prediction lines.
- When w = 50, the model underestimates housing prices.
- When w = 100, the predictions improve but still do not match the actual values.
- When w = 200, the prediction line passes exactly through the training examples.
- Therefore, w = 200 gives the minimum cost and is the optimal parameter for this dataset.
